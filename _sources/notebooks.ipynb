## Data Nominal/Categorical
* Nilai atribut nominal adalah simbol ataunama barang. Setiap nilai mewakili beberapa jenis kategori, kode, atau status, dan Atribut nominal juga disebut kategori. Nilai-nilainya tidak memiliki tingkatan nilai. Dalam ilmu komputer, nilainya juga dikenal sebagai enumerasi.
* Melakukan pengukuran terhadap data nominal dengan menggukan rumus : 
$$ d(i,j)=\frac{p - m}p $$
$ p $ = jumlah data yang akan dihitung jaraknya<br>
$ m $ = jumlah data yang sama


```python
# Menampilkan data Nominal/Categorical
mobile[["Mobile_Phone","Model","Review", "Payment_Method"]].head()
```

Melakukan perhitungan berdasarkan pada data nominal

```python
# Menghitung jarak data Nominal/categorical
def d1(i,j):
    P = 4
    M = 0
    for col in ["Mobile_Phone", "Model", "Review","Payment_Method"]:
        if mobile[col][i] == mobile[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan jarak di tipe data nominal/categorical
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

## Data Numeric
* Atribut numerik bersifat kuantitatif; Artinya, ini adalah kuantitas yang terukur, yang dinyatakan dengan bilangan bulat atau nilai riel. Atribut numerik dapat Atribut Skala Interval(interval-scaled) atau skala ration (ratio-scaled)
* Melakukan pengukuran terhadap data nominal dengan menggukan rumus Euclian Distance:
$$ d(i,j) = \sqrt(|{x_{i1} - x_{j1}}|^2 + |{x_{i2} - x_{j2}}|^2 + ... + |{x_{ip} - x_{jp}}|^2) $$

```python
# Menampilkan data numeric
mobile[["RAM_GB", "ROM_GB", "Latest_Price_Taka", "Discount", "Selling_Price"]].head()
```

```python
# Melihat detail dari data numeric yang ada
mobile.describe()
```

Karena datanya tidak konsisten maka, data perlu dilakukan normalisasi dengan menggunakan rumus : 
$$ z=\frac{x - u}o $$
$ x $ : Data <br>
$ u $ : Rata-rata <br>
$ o $ : Standar deviasi

```python
# Normalisasi data RAM_GB
std1 = 2.920372
mean1 = 5.142857
for col1 in ["RAM_GB"]:
    data1 = mobile[col1]
    hasil1 = (data1-mean1)/std1
    print(hasil1)

# Normalisasi data ROM_GB
std2 = 62.026722
mean2 = 93.714286
for col2 in ["ROM_GB"]:
    data2 = mobile[col2]
    hasil2 = (data2-mean2)/std2
    print(hasil2)

# Normalisasi data Latest_Price_Taka
std3 = 36891.663754	
mean3 = 37610.952381	
for col3 in ["Latest_Price_Taka"]:
    data3 = mobile[col3]
    hasil3 = (data3-mean3)/std3
    print(hasil3)

# Normalisasi data Discount
std4 = 766.097097
mean4 = 1940.476190	
for col4 in ["Discount"]:
    data4 = mobile[col4]
    hasil4 = (data4-mean4)/std4
    print(hasil4)

# Normalisasi data Selling_Price
std5 = 36284.165145
mean5 = 35670.476190
for col5 in ["Selling_Price"]:
    data5 = mobile[col5]
    hasil5 = (data5-mean5)/std5
    print(hasil5)
```

Melakukan perhitungan berdasarkan rumus Euclidean Distance
$$ d(i,j) = \sqrt(|{x_{i1} - x_{j1}}|^2 + |{x_{i2} - x_{j2}}|^2 + ... + |{x_{ip} - x_{jp}}|^2) $$

```python
# Menghitung jarak data Numeric dengan cara Euclidean Distance
def d2(i,j,h):
    result = 0
    for col in [hasil1, hasil2, hasil3, hasil4, hasil5]:
        result = math.pow(abs(col[i] - col[j]), h)
    return result**(1/h)
```

```python
# Menampilkan hasil perhitungan dari data numeric
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

## Equal Width

Rumus perhitungan yang digunakan menggunakan **Equal Width** yaitu : 
$$ w=\frac{x_{min}-x_{max}}k $$

$$v_i=x_{min}+i*w$$
$ x $ = nilai <br>
$ i $ = 1,2,3,.. <br>
$ k $ = jumlah penglompokan

```python
# Proses diskritisasi sepal_length
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=4,
                       encode='ordinal', 
                       strategy='uniform')
iris["sepal_length"] = est.fit_transform(iris[["sepal_length"]])
```

## Model Klasifikasi 

```python
#Now try to create a model to solve our task
#As per our analysis, we can't find much information from new feature which can helpful in solving our problem...
#For solving our task I have selected few features amongs all to build up our best model..

'''Imporing few library for create Decision tree classifier and visualizing the tree structure'''

from sklearn import tree
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score


'''Here we separating independent varibles or target varibles from Iris dataset'''


X = Iris_data[['SepalLengthCm', 'SepalWidthCm','PetalLengthCm', 'PetalWidthCm','Sepal_petal_wid_len_diff','Sepal_petal_width_diff']]
y = Iris_data['Species']


#Before training the model we have split our data into Actual Train and Actual Test Dataset for training and validating purpose...

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.30, random_state=42)

#spliting data into validation train and validation test
Xt, Xcv, Yt, Ycv = train_test_split(Xtrain, Ytrain, test_size=0.10, random_state=42)


'''Now we have create a Decision tree classifier and trained it with training dataset.'''


Iris_clf = DecisionTreeClassifier(criterion='gini',min_samples_split=2)
Iris_clf.fit(Xt, Yt)

#Visualized the Tree which is formed on train dataset

tree.plot_tree(Iris_clf)
```

```python
#Visualizing Decision Tree using graphviz library

dot_data = tree.export_graphviz(Iris_clf, out_file=None)

graph = graphviz.Source(dot_data)
graph
```

```python
# As our model has been trained....
#Now we can validate our Decision tree using cross validation method to get the accuracy or performance score of our model.

print('Accuracy score is:',cross_val_score(Iris_clf, Xt, Yt, cv=3, scoring='accuracy').mean())
```

```python
#Checking validation test data on our trained model and getting performance metrices

from sklearn.metrics import multilabel_confusion_matrix, accuracy_score

Y_hat = Iris_clf.predict(Xcv)


print('Accuracy score for validation test data is:',accuracy_score(Ycv, Y_hat))
multilabel_confusion_matrix(Ycv , Y_hat)
```

```python
#Checking our model performance on actual unseen test data.. 
YT_hat = Iris_clf.predict(Xtest)
YT_hat

print('Model Accuracy Score on totally unseen data(Xtest) is:',accuracy_score(Ytest, YT_hat)*100,'%')
multilabel_confusion_matrix(Ytest , YT_hat)
```

* Seperti yang kami ketahui, fitur yang kami pilih berfungsi dengan baik dan model memberikan skor akurasi yang sangat baik pada validasi atau data uji aktual. Jadi Sekarang kita dapat melatih model kita pada dataset kereta Aktual dengan fitur yang dipilih untuk mengevaluasi/menerapkan model kita dalam kasus dunia nyata.

```python
'''Training model on Actual train data... '''
Iris_Fclf = DecisionTreeClassifier(criterion='gini',min_samples_split=2)
Iris_Fclf.fit(Xtrain, Ytrain)

#Visualize tree structure..
tree.plot_tree(Iris_Fclf)
```

```python
#Final Decision tree build for deploying in real world cases....

dot_data = tree.export_graphviz(Iris_Fclf, out_file=None)
graph = graphviz.Source(dot_data)
graph
```

```python
#Checking the performance of model on Actual Test data...

YT_Fhat = Iris_Fclf.predict(Xtest)
YT_Fhat

print('Model Accuracy Score on totally unseen data(Xtest) is:',accuracy_score(Ytest, YT_Fhat)*100,'%')
multilabel_confusion_matrix(Ytest , YT_Fhat)
```
