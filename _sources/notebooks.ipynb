```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

N = 10
data = [np.logspace(0, 1, 100) + np.random.randn(100) + ii for ii in range(N)]
data = np.array(data).T
cmap = plt.cm.coolwarm
rcParams['axes.prop_cycle'] = cycler(color=cmap(np.linspace(0, 1, N)))


from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color=cmap(0.), lw=4),
                Line2D([0], [0], color=cmap(.5), lw=4),
                Line2D([0], [0], color=cmap(1.), lw=4)]

fig, ax = plt.subplots(figsize=(10, 5))
lines = ax.plot(data)
ax.legend(custom_lines, ['Cold', 'Medium', 'Hot']);
```

```python
# Import liberary yang diperlukan
import pandas as pd 
import math
```

```python
# Membaca data csv
mobile = pd.read_csv("https://raw.githubusercontent.com/Luqman-Noor-Buat/datamining/main/Samsung_mobile.csv")
mobile.head()
```

```python
# mengetahui jumlah baris dan kolom
mobile.shape
```

Data Nominal/Categorical

```python
# Menampilkan data Nominal/Categorical
mobile[["Mobile_Phone","Model","Review", "Payment_Method"]].head()
```

There is a lot more that you can do with outputs (such as including interactive outputs)
with your book. For more information about this, see [the Jupyter Book documentation](https://jupyterbook.org)

```python
# Menghitung jarak data Nominal/categorical
def d1(i,j):
    P = 4
    M = 0
    for col in ["Mobile_Phone", "Model", "Review","Payment_Method"]:
        if mobile[col][i] == mobile[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

Data Numeric

```python
# Menampilkan data numeric
mobile[["RAM_GB", "ROM_GB", "Latest_Price_Taka", "Discount", "Selling_Price"]].head()
```

```python
# Menghitung jarak data Numeric dengan cara Euclidean
def d2(i,j,h):
    result = 0
    for col in ["RAM_GB", "ROM_GB", "Latest_Price_Taka", "Discount", "Selling_Price"]:
        result = math.pow(abs(mobile[col][i] - mobile[col][j]), h)
    return result**(1/h)
```

```python
# Menampilkan hasil perhitungan
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

Similarity

```python
# Jarak d(1,2)
print(d1(1,2)+d2(1,2,2))
```

```python
# Jarak d(1,3)
print(d1(1,3)+d2(1,3,2))
```

```python
# Jarak d(1,4)
print(d1(1,4)+d2(1,4,2))
```
