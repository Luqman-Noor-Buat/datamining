```python
# Import liberary yang diperlukan
import pandas as pd 
import math
```

```python
# Membaca data csv
mobile = pd.read_csv("https://raw.githubusercontent.com/Luqman-Noor-Buat/datamining/main/Samsung_mobile.csv")
mobile.head()
```

```python
# mengetahui jumlah baris dan kolom
mobile.shape
```

Data Nominal/Categorical

```python
# Menampilkan data Nominal/Categorical
mobile[["Mobile_Phone","Model","Review", "Payment_Method"]].head()
```

```python
# Menghitung jarak data Nominal/categorical
def d1(i,j):
    P = 4
    M = 0
    for col in ["Mobile_Phone", "Model", "Review","Payment_Method"]:
        if mobile[col][i] == mobile[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

Data Numeric

```python
# Menampilkan data numeric
mobile[["RAM_GB", "ROM_GB", "Latest_Price_Taka", "Discount", "Selling_Price"]].head()
```

```python
import statistics
# Mencari rata-rata RAM_GB
print(statistics.mean(mobile["RAM_GB"]))

# Mencari rata-rata ROM_GB

# Mencari rata-rata Latest_Price_Taka

# Mencari rata-rata Discount

# Mencari rata-rata Selling_Price
```

```python
# Mencari standart deviasi RAM_GB
print(statistics.stdev(mobile["RAM_GB"]))

# Mencari standart deviasi ROM_GB

# Mencari standart deviasi Latest_Price_Taka

# Mencari standart deviasi Discount

# Mencari standart deviasi Selling_Price
```

```python
# Menghitung jarak data Numeric dengan cara Euclidean
def d2(i,j,h):
    result = 0
    for col in ["RAM_GB", "ROM_GB", "Latest_Price_Taka", "Discount", "Selling_Price"]:
        result = math.pow(abs(mobile[col][i] - mobile[col][j]), h)
    return result**(1/h)
```

```python
# Menampilkan hasil perhitungan
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

Similarity

```python
# Jarak d(1,2)
print(d1(1,2)+d2(1,2,2))
```

```python
# Jarak d(1,3)
print(d1(1,3)+d2(1,3,2))
```

```python
# Jarak d(1,4)
print(d1(1,4)+d2(1,4,2))
```
